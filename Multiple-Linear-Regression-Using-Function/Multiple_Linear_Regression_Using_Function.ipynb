```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D as ax
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Linear Regression/winequality-red.csv')
```

```python
features_to_select = ['fixed acidity', 'volatile acidity', 'quality']
X = data[features_to_select]
```

```python
data.head(9)
```

```python
plt.scatter(data['fixed acidity'], data['quality'],  color='blue')
plt.xlabel("fixed acidity")
plt.ylabel("quality")
plt.show()
```

```python
msk = np.random.rand(len(data)) < 0.8
train = data[msk]
test = data[~msk]
```

```python
plt.scatter(train['fixed acidity'], train['quality'],  color='blue')
plt.xlabel("Fixed Acidity")
plt.ylabel("Quality")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['fixed acidity','volatile acidity']])
y = np.asanyarray(train[['quality']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['fixed acidity','volatile acidity']])
x = np.asanyarray(test[['fixed acidity','volatile acidity']])
y = np.asanyarray(test[['quality']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
